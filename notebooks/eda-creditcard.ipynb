# Exploratory Data Analysis – Credit Card Fraud Dataset

This notebook explores the credit card transaction dataset used for fraud detection.
The goal is to understand data structure, clean the dataset, analyze class imbalance,
and extract insights relevant for modeling.

```python
print("Importing all the dependencies needed...")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
print("Imported dependencies")
```

```python
print("Loading the dataset")
credit_df = pd.read_csv("./data/raw/creditcard.csv")
```

```python
print("Initial inspection")
print("The features V1–V28 are anonymized principal components obtained via PCA As PCA centers data around zero, both positive and negative values are expected and represent deviations from the mean along each component axis.These values do not indicate errors or anomalies but encode latent transaction patterns relevant for fraud detection")
credit_df.head()
```

```python
print("Shape of the dataset")
credit_df.shape
```

### Dataset Overview

- Each row represents a credit card transaction
- Features `V1`–`V28` are anonymized PCA components
- `Time` represents seconds since the first transaction
- `Amount` is the transaction value
- `Class` is the target variable:
  - `1` → Fraud
  - `0` → Legitimate

```python
credit_df.info()
```

### Missing value check

The dataset contains no missing values. This is expected as the data has been preprocessed and anonymized.


```python
credit_df.duplicated().sum()
```

```python
credit_df = credit_df.drop_duplicates()
credit_df.duplicated().sum()
```

### Duplicate Removal

Duplicate transactions were removed to avoid biasing model learning.
This ensures each transaction contributes only once.

```python
print("Statisical summary")
credit_df.describe()
```

### Observations
- PCA features are centered around zero
- `Amount` is highly right-skewed
- `Time` spans a long transaction duration
- Feature scaling will be required before modeling

## Exolaritary Data Analysis (EDA)

## Class Distribution Analysis

Fraud detection datasets are typically highly imbalanced.
This section quantifies the imbalance in the target variable.

```python
print("Class counts...")
credit_df["Class"].value_counts()
```

```python
print("Class percentage")
credit_df["Class"].value_counts(normalize=True) * 100
```

```python
print("Visualize class distribution")
sns.countplot(x="Class", data=credit_df)
plt.title("Class Distribution (Fraud vs Non-Fraud)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

### Class Imbalance Insight

- Fraud cases represent less than 1% of all transactions
- Accuracy alone is misleading for evaluation
- Precision, Recall, F1-score, and AUC-PR will be required
- Resampling techniques will be applied only on training data

## Univariate Analysis

This section examines the distribution of key numerical features.

```python
print("Transaction amount distribution")
plt.figure(figsize=(8, 4))
sns.histplot(credit_df["Amount"], bins=50, kde=True)
plt.title("Distribution of Transaction Amount")
plt.xlabel("Amount")
plt.show()
```

### Amount Distribution Insight

- Transaction amounts are heavily right-skewed
- Most transactions are of small value
- Log transformation or scaling will be beneficial

```python
print("Time distribution")
plt.figure(figsize=(8, 4))
sns.histplot(credit_df["Time"], bins=50)
plt.title("Distribution of Transaction Time")
plt.xlabel("Time (seconds)")
plt.show()
```

### Time Feature Insight

- Transactions are unevenly distributed over time
- Feature engineering could extract hour-of-day signals

## Bivariate Analysis

This section explores relationships between features and fraud labels.

```python
print("Amount VS Class")
plt.figure(figsize=(8, 4))
sns.boxplot(x="Class", y="Amount", data=credit_df)
plt.yscale("log")
plt.title("Transaction Amount by Class")
plt.show()
```

### Amount vs Fraud Insight

- Fraudulent transactions often involve higher amounts
- However, some fraud occurs at low amounts
- Amount alone is insufficient for detection

```python
print("Correlation with target")
corr = credit_df.corr()["Class"].sort_values(ascending=False)
corr.head(10)
```

### Correlation Analysis Insight

- Certain PCA components show strong correlation with fraud
- These features capture hidden fraud patterns
- Feature importance analysis will later validate these signals
